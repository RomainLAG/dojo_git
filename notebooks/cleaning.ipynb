```python
import pandas as pd
```

```python
df_customers = pd.read_csv("https://dojo-git.s3.eu-west-3.amazonaws.com/raw/customers.csv")
df_transactions = pd.read_csv("https://dojo-git.s3.eu-west-3.amazonaws.com/raw/transactions.csv")
df_products = pd.read_csv("https://dojo-git.s3.eu-west-3.amazonaws.com/raw/products.csv")
```

```python
print(df_products.isnull().sum())
print(df_transactions.isnull().sum())
print(df_customers.isnull().sum())
```

```python
df_transactions['date'] = pd.to_datetime(df_transactions['date'],infer_datetime_format=True, errors='coerce')
df_transactions.head(20)
```

```python
df_transactions.drop_duplicates(inplace=True)
df_customers.drop_duplicates(inplace=True)
df_products.drop_duplicates(inplace=True)
```

```python
df_customers.rename(columns={'client_id': 'client_id', 'sex': 'gender', 'birth': 'birth_year'}, inplace=True)
df_transactions.rename(columns={'id_prod': 'product_id', 'date': 'purchase_date', 'session_id': 'session_id', 'client_id': 'client_id'}, inplace=True)
df_products.rename(columns={'id_prod': 'product_id', 'price': 'price', 'categ': 'category'}, inplace=True)
```
